```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
ns = np.array([10, 100, 1000, 10448, 101148, 506516, 2462922])
rts = np.array([61, 335, 1682, 18617, 172807, 885401, 4455458])
coef1 = np.polyfit(ns, rts / 1000, 1) # get linear regression coefficients
poly1d_fn1 = np.poly1d(coef1)
plt.figure(figsize=(6,4))
plt.plot(ns, rts / 1000, '*')
plt.plot(ns, poly1d_fn1(ns), '--c')
plt.title('Runtime of searchKMP algorithm with various data sizes (100 runs)')
plt.ylabel(r'Runtime of 100 runs (ms)')
plt.xlabel('Characters in dataset (n)')
plt.annotate(r'$R^2$ = {:.3f}'.format(r2_score(rts / 1000, poly1d_fn1(ns))), (1.5e6, 2300))
plt.figure(figsize=(6,4))
plt.title('Runtime of searchKMP with various data sizes log-log plot (100 runs)')
plt.ylabel(r'Runtime of 100 runs (ms)')
plt.xlabel('Characters in dataset (n)')
plt.loglog(ns, rts / 1000, '.')
```
